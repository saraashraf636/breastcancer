```python
import matplotlib.pyplot as plt
import numpy as np

thickness=[5,3,6,4,8]
size=[4,1,8,1,10]
plt.plot(thickness,size,label='2x', color='red' ,linewidth=2 ,marker='.',markersize=10,markeredgecolor='blue')
plt.title(' the relation between cellthickness and cellsize',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('cellthickness')
plt.ylabel('cellsize')
plt.legend()
plt.show()
```
